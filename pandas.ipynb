```python
plt.rcParams['figure.figsize'] = 20, 10
vis3 = sns.boxplot(stats, x='income_group', y='birth_rate', legend=['Internet users', 'Country Code'])

# Calculate IQR and identify outliers for each income group
for income_group in stats['income_group'].unique():
    group_data = stats[stats['income_group'] == income_group]
    Q1 = group_data['birth_rate'].quantile(0.25)
    Q3 = group_data['birth_rate'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Outliers
    outliers = group_data[(group_data['birth_rate'] < lower_bound) | (group_data['birth_rate'] > upper_bound)]

    # Annotate outliers with country code
    for _, outlier in outliers.iterrows():
        plt.annotate(outlier['country_name'],
                     (outlier['income_group'], outlier['birth_rate']),
                     textcoords="offset points",  # how to position the text
                     xytext=(0, 10),  # distance from text to points (x,y)
                     ha='center')  # horizontal alignment can be left, right or center

plt.show()
```
