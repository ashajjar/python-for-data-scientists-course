```python
def my_plot(data):
    markers = ["d", "v", "s", "*", "^", "d", "v", "s", "*", "^"]

    for player in Players:
        plt.plot(data[Pdict[player]], marker=markers[Pdict[player]], label=player)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.xticks(list(range(0, 10)), Seasons, rotation='vertical')
    plt.xlabel('Season')
    plt.ylabel("Player's Data")
    plt.show()
```
